```python
import getAreas as geta
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# example heightmap of the Rhone Valley
hm = geta.Heightmap((45.010755, 5.492421), (45.350255, 6.220719), 100, 100)
# VÃ©rhalom:
#hm = geta.Heightmap((47.507439, 18.982058), (47.526874, 19.015201), 100, 100)
```

```python
hm.getElevations()
hm.normalizeElevations()
hm.exportHeightmap(outName="heights")
```

```python
# Visualization:
elevs = np.array(hm.elevations)
elevs = elevs.reshape(100,100)
X = np.arange(0,100,1)
Y = np.arange(0,100,1)
X_mesh, Y_mesh = np.meshgrid(X, Y)

fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X_mesh, Y_mesh, elevs, cmap='viridis')
ax1.set_title('3D plot')

ax2 = fig.add_subplot(122)
contour = ax2.contourf(X_mesh, Y_mesh, elevs, cmap = 'viridis')
ax2.set_title('X-Y plane')

cbar = plt.colorbar(contour)

plt.tight_layout()
plt.show()
```
